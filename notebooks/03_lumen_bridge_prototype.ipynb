# Lumen bridge prototype

Inspect how AtmosLens exposes the xarray schema and serializes an analysis request as pipeline-like transform steps.

```python
from atmoslens.datasets import load_dataset
from atmoslens.lumen_bridge import XarrayPipelineBridge
from atmoslens.models import AnalysisRequest
from atmoslens.recommendations import activity_pipeline_steps

ds = load_dataset("../data/sample_forecast.nc")
bridge = XarrayPipelineBridge(ds)
request = AnalysisRequest(
    location_name="Docklands",
    profile_name="Sensitive",
    activity_name="Run",
    pollutant="ozone",
    advisor_mode="Next 24 hours",
)
bridge.schema()
```

```python
bridge.build_query_spec(request, activity_pipeline_steps(request))
```
