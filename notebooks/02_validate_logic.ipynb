# Validate recommendation logic

Exercise the same scoring and recommendation pipeline used in the Panel app.

```python
from atmoslens.datasets import load_dataset
from atmoslens.models import AnalysisRequest
from atmoslens.recommendations import build_activity_result, build_route_result

ds = load_dataset("../data/sample_forecast.nc")
request = AnalysisRequest(
    location_name="Docklands",
    profile_name="Sensitive",
    activity_name="Run",
    pollutant="ozone",
    advisor_mode="Next 24 hours",
)
result = build_activity_result(ds, request)
result.recommendation
```

```python
route_request = AnalysisRequest(
    location_name="Docklands",
    profile_name="Sensitive",
    activity_name="Cycle Commute",
    pollutant="ozone",
    advisor_mode="Next 24 hours",
    route_name="Sandyford to Docklands",
)
build_route_result(ds, route_request).recommendation
```
